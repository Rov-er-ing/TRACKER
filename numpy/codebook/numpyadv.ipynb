```python
import numpy as np
```

```python
#fancy indexing 
a=np.arange(24).reshape(6,4)
print(a)

#getting 1st 3rd nd 4th row directly
a[[0,2,3]]
a[:,[1,2]]#colum 2 and 3
a[np.ix_([0,2,3],[1,2])]# 1,3,4 row with 2,3 column
```

```python
#boolean iindexing
a=np.random.randint(1,100,24).reshape(6,4)
a>50#boolean array
a[a>50]#gives only true value

#find even numbers
a[a%2==0]

#find no greater than 50 and even 
a[(a>50)& (a%2==0)]
#not divisible by 7
a[~(a%7==0)]
```

 Broadcasting is a mechanism that allows NumPy to perform operations on arrays of different shapes.

 rule1:make two array have the same no of dim
       . if the number of dimension of the two array are different , add new dimension with size 1 to head of array with smaller dim
       a(3,3,3),b(2) then b broadcast then b(1,1,2)

 rule2:make each dim of two array same size
       . if size of each dim of two array do not match,dim with size 1 aare stretched to the size of other array
       . if there is dim whose size is not 1 in either of two array,it cannot broadcast and an error is raised

```python
a=np.arange(3).reshape(1,3)
b=np.arange(3).reshape(3,1)
print(a)
print(b)

a+b#broadcast
  

```

```python

a=np.arange(12).reshape(3,4)
b=np.arange(3)
print(a)
print(b)
a+b #not broadcast
when (1,3) strech to (3,3) still not equal to (3,4)
```

```python
#sigmoid
def sigmoid(x):
    return(1/(1+np.exp(-(x))))
b=np.arange(5)
sigmoid(b)
```

```python
#mean square error
def mse(a,p):
   return np.mean((a-p)**2)

a=np.random.randint(1,50,25)
p=np.random.randint(1,50,25)
mse(a,p)
```

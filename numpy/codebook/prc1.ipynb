```python
#softmax function
#exp(x)/sum(exp(x))
def softmax(x):
    p=np.exp(x)
    q=sum(p)
    return p/q
x=[1,2,3,4]
print(softmax(x))
```

```python
def date(start,end):
   y1,d1,m1=map(int,start.split("-"))
   y2,d2,m2=map(int,end.split("-"))
   if y1!=y2:
      print("year arnt same")
   else:
      
    s=np.datetime64(start)
    d=np.datetime64(end)
    return np.arange(s,d+1,dtype='datetime64[D]') 
   

print(date("2022-01-01","2022-01-10"))
      
      
```

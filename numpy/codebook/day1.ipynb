stacking and spliting  of array 
1. horizontal stacking
 2x2 + 2x2 = 2x4(col inc)

2. vertical stacking 
2x2+ 2x2= 4x2(row inc)

3. horizontal spliting 
 divided the coln in n equal parts

4.vertical split
divide the row into n equal parts


```python
#horizontal
#np.hstack((a1,a2,a3..))
a1=np.arange(9).reshape(3,3)
a2=np.arange(9,18).reshape(3,3)
a3=np.hstack((a1,a2))

#vertical stack
#np.vstack((a1,a2,a3..))
a4=np.vstack((a1,a2))

#hsplit
#print(a3)
np.hsplit(a3,3)

#vsplit
print(a4)
np.vsplit(a4,3)
```
